# OpenPlazma Workbench Lite

This notebook uses `STATIC_FIXTURE` data only. It is not a validated fusion simulation and it is not a real hardware experiment.

```python
import openplazma_lite as op

ctx = op.load_context()
ctx
```

```python
signal_id = ctx["signals"][0]["signalId"]
signal_path = "signals/plasma_current.json"
signal = op.load_signal(signal_path)
assert signal["signalId"] == signal_id
op.plot_signal(signal, time_range=ctx.get("view", {}).get("timeRange"))
```

```python
observations = ctx.get("observations", [])
hypothesis = ctx.get("hypothesis", "No hypothesis provided.")
print("Observations:")
for observation in observations:
    print("-", observation.get("text", observation))
print("Hypothesis:", hypothesis)
```
