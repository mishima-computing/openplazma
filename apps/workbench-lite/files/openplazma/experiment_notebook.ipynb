# OpenPlazma Workbench Lite

This notebook uses `STATIC_FIXTURE` data only. It is a read-only analysis and decision-support example: it helps inspect evidence, assumptions, and limitations without controlling equipment or providing hazardous procedures.

```python
import openplazma_lite as op

ctx = op.load_context()
assert ctx["source"]["provider"] == "STATIC_FIXTURE"
assert ctx["capabilities"]["controlFacility"] is False
ctx
```

```python
signal_id = ctx["signals"][0]["signalId"]
signal_path = "signals/plasma_current.json"
signal = op.load_signal(signal_path)
assert signal["signalId"] == signal_id
op.plot_signal(signal, time_range=ctx.get("view", {}).get("timeRange"))
```

```python
observations = ctx.get("observations", [])
hypothesis = ctx.get("hypothesis", "No hypothesis provided.")
print("Observations:")
for observation in observations:
    print("-", observation.get("text", observation))
print("Hypothesis:", hypothesis)
```

Limitations:

- This notebook uses STATIC_FIXTURE data only.
- This is read-only analysis and decision support.
- This notebook has no command/control path or hazardous operating procedure.
- This notebook is not a standalone authority for safety-critical operation or reactor design decisions.
